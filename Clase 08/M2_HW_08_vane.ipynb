```python
def consulta_sql(consulta):
    cursor.execute(consulta)
    resultados = cursor.fetchall()
    # convertimos a df trayendo los nombres de las columnas
    df = pd.DataFrame(resultados, columns = [desc[0] for desc in cursor.description])
    return df
```

3. A partir de tener los datos disponibles, responder a las siguientes preguntas:

a) ¿Cuál es el barrio con mayor cantidad de Pubs?


```python
df_filtrado.categoria.unique()
```

```python
consulta = """
SELECT barrio, categoria, count(idLocal) as cantidad
FROM ofertaGastronomica
GROUP BY barrio, categoria
HAVING categoria = 'PUB'
ORDER BY cantidad DESC
LIMIT 1
"""
consulta_sql(consulta)
```

```python
consulta = """
SELECT barrio, categoria, count(idLocal) as cantidad
FROM ofertaGastronomica
WHERE categoria = 'PUB'
GROUP BY barrio, categoria
ORDER BY cantidad DESC
LIMIT 1
"""
consulta_sql(consulta)
```

   b) Obtener la cantidad de locales por categoría


```python
consulta = """
SELECT categoria, count(idLocal) as cantidad
FROM ofertaGastronomica
GROUP BY categoria
ORDER BY cantidad DESC
"""
consulta_sql(consulta)
```

   c) Obtener la cantidad de restaurantes por comuna

```python
consulta = """
SELECT comuna, categoria, count(idLocal) as cantidad
FROM ofertaGastronomica
WHERE categoria = 'RESTAURANTE'
GROUP BY comuna
ORDER BY cantidad DESC
"""
consulta_sql(consulta)
```
